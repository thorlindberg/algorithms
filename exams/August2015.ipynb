```
def classifyBMI(height, weight):

  b = weight / height**2

  if b < 16:
    BMIGroup = 'Severely underweight'
  if 16 <= b < 18.5:
    BMIGroup = 'Underweight'
  if 18.5 <= b < 25:
    BMIGroup = 'Normal'
  if 25 <= b < 30:
    BMIGroup = 'Overweight'
  if 30 <= b < 40:
    BMIGroup = 'Obese'
  if 40 <= b:
    BMIGroup = 'Severely obese'

  return BMIGroup
```

```
def cumulativeStock(transactions):

  import numpy as np

  items = 0
  stock = []
  for transaction in transactions:
    if transaction == 0:
      items = 0
    if transaction > 0:
      items += transaction
    if transaction < 0:
      items += transaction
    stock.append(items)
  stock = np.array(stock)

  return stock
```

```
def RPNCalculator(commands):

  import numpy as np

  stack = []
  commands = commands.split()
  for command in commands:
    if command not in ['+', '-', 's']:
      stack.append(int(command))
    if command == '+':
      addition = stack[-2] + stack[-1]
      stack = stack[0:-2]
      stack.append(addition)
    if command == '-':
      subtraction = stack[-2] - stack[-1]
      stack = stack[0:-2]
      stack.append(subtraction)
    if command == 's':
      stack[-1], stack[-2] = stack[-2], stack[-1]
  stack = np.array(stack)

  return stack
```

```
def starPoints(a, b, maxArea):

  import math

  area = 0
  points = 0
  while area < maxArea:
    points += 1
    area = points * a * b * math.sin(math.pi / points)
  
  if area > maxArea:
    points -= 1

  return points
```

```
def matrixCleanup(M):

  import numpy as np

  columns = []
  for column in range(len(M[0])):
    if 0 in M[:,column]:
      columns.append(column)
  
  MClean = []
  for row in M:
    newrow = []
    for index in range(len(row)):
      if index not in columns and 0 not in row:
        newrow.append(row[index])
    if newrow != []:
      MClean.append(newrow)
  MClean = np.array(MClean)

  return MClean
```
