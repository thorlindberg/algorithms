```
def trianglearea(a, b, c):

  import math

  if a + b > c and a + c > b and b + c > a:
    p = 0.5 * (a + b + c)
    A = math.sqrt(p * (p - a) * (p - b) * (p - c))
  else:
    return 0

  return A
```

```
def polygonarea(x, y):

  import numpy as np

  formula = 0
  for index in range(len(x) - 1):
    formula += np.linalg.det(np.array([[x[index], x[index+1]], [y[index], y[index+1]]]))
  formula += np.linalg.det(np.array([[x[-1], x[0]], [y[-1], y[0]]]))
  
  A = 0.5 * formula

  return A
```

```
def predictability(x):

  import numpy as np

  q = np.array([])

  for number in np.unique(x):
    q = np.append(q, x.count(number))

  return q
```

```
def syllables(word):

  vowels = ['a', 'e', 'i', 'o', 'u', 'y']
  
  word = word[::-1]
  index = -1
  for letter in word:
    index += 1
    if letter not in vowels:
      word = word[index:]
      break
  word = word[::-1]

  N = 0
  style = ''
  for letter in word:
    if letter in vowels and style != '+':
      style = '+'
      N +=1
    if letter not in vowels:
      style = '-'
  
  return N
```
