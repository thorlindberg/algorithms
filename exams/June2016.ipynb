```
def survivalTemperature(M,g):

  import numpy as np

  if np.min(M) < 50 or 500 < np.max(M):
    return 'RangeError'
  if np.min(g) < 0.04 or 0.45 < np.max(g):
    return 'RangeError'

  T = []
  for m in M:
    row = []
    for value in g:
      row.append(36 - ((0.9*m-12)*(value+0.95))/(27.8*value))
    T.append(row)
  T = np.array(T)

  return T
```

```
def greenEyes(p1,p2):

  chart = {
  'brownbrown' : 19,
  'brownblue' : 0,
  'browngreen' : 38,
  'blueblue' : 1,
  'bluegreen' : 50,
  'greengreen' : 75
  }

  if p1+p2 in chart:
    P = chart[p1+p2]
  if p2+p1 in chart:
    P = chart[p2+p1]

  return P
```

```
def bestBuy(p,m):

  n = 0
  spent = 0
  for price in p:
    if spent + price <= m:
      spent += price
      n += 1
    else:
      break

  return n
```
