```
def nBest(allValues, N):

  import numpy as np

  sort = np.sort(allValues)[::-1][:N]

  bestValues = []
  for value in allValues:
    if value in sort:
      bestValues.append(value)
  bestValues = np.array(bestValues)

  return bestValues
```

```
def timeDilation(t, v):

  import numpy as np
  import math

  T = []
  for time in t:
    row = []
    for velocity in v:
      row.append(time / (math.sqrt(1 - velocity**2 / 299792458**2)))
    T.append(row)
  T = np.array(T)

  return T
```

```
def derangements(n):

  import math

  k = math.factorial(n) / math.e
  if abs(math.ceil(k) - math.factorial(n) / math.e) <= abs(math.floor(k) - math.factorial(n) / math.e):
    k = math.ceil(k)
  else:
    k = math.floor(k)

  return k
```

```
def industry(industryCode):

  industries = {
  '0000-0999' : 'Agriculture',
  '1000-1499' : 'Mining',
  '1500-1999' : 'Construction',
  '2000-3999' : 'Manufacturing',
  '4000-4999' : 'Transportation',
  '5000-5999' : 'Trade',
  '6000-6999' : 'Finance',
  '7000-8999' : 'Services',
  '9000-9999' : 'Public'
  }

  for industry in industries:
    for digit in range(10):
      code = industryCode.replace('x', str(digit))
      if int(industry[:4]) <= int(code) <= int(industry[5:9]):
        industryName = industries[industry]
        return industryName
```

```
def taskDispatch(tasks, units):

  import numpy as np

  assignment = np.zeros(np.size(tasks))
  for unit in range(len(units)):
    for task in range(len(tasks)):
      if assignment[task] == 0:
        if units[unit] >= tasks[task]:
          units[unit] = units[unit] - tasks[task]
          assignment[task] = unit + 1

  return assignment
```
