```
def weeklyAverage(dayTemp):

  import numpy as np

  weekTemp = np.mean(np.sort(dayTemp)[2:-2])

  return weekTemp
```

```
def carsAvailable(fleet, category):
  
  categories = {
  'm' : fleet[0],
  'e' : fleet[1],
  'c' : fleet[2],
  's' : fleet[3],
  'f' : fleet[4],
  'l' : fleet[5],
  }

  if category[0].lower() not in 'mecsfl':
    n = -1
  else:
    n = categories[category[0].lower()]
  
  return n
```

```
def speedOfLight(f, wavelengths):

  import numpy as np

  lengths = []
  for wave in wavelengths:
    lengths.append(wave * f)
  lengths = np.array(lengths)
  c = np.mean(lengths)

  return c
```

```
def similarityMatrix(x, y, delta):

  import numpy as np
  import math
  
  S = []
  for i in x:
    row = []
    for j in y:
      row.append(math.exp(-delta * (i - j)**2))
    S.append(row)

  S = np.array(S)

  return S
```
