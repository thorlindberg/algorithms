```
def planets(planetCategories):

  import numpy as np

  categories = {
  'Mercury' : [1, 2],
  'Venus' : [1, 2],
  'Earth' : [1],
  'Mars' : [1, 3],
  'Jupiter' : [3, 4, 5],
  'Saturn' : [3, 4, 5],
  'Uranus' : [3, 4, 5],
  'Neptune' : [3, 4, 5],
  'Pluto' : [3, 4]
  }

  planetNames = np.array([])

  for category in categories:
    for number in planetCategories:
      if number in categories[category]:
        if category not in planetNames:
          planetNames = np.append(planetNames, category)

  return planetNames
```

```
def inversions(board):

  inv = 0
  for index in range(len(board) - 1):
    if board[index] > board[index + 1]:
      inv += 1

  return inv
```

```
def polygon(x, y):

  import numpy as np

  C = np.array([])

  for index in range(len(x) - 2):
    c = np.dot(x[index] - x[-1], y[index + 1] - y[index]) - np.dot(x[index + 2] - x[index], y[index] - y[-1])
    C = np.append(C, c)

  return C
```

```
def polygon(x, y):

  import numpy as np

  n = np.size(x)
  C = np.zeros(n)
  for k in range(n):
    if k==0:
      kminus = n-1
    else:
      kminus = k-1
    if k==n-1:
      kplus = 0
    else:
      kplus = k+1

    C[k] = (x[k]-x[kminus])*(y[kplus]-y[k]) - (x[kplus]-x[k])*(y[k]-y[kminus])

  return C
```
