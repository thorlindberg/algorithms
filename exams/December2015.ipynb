```
def convertGrade(grade, scale):

  scale13 = {
  '13' : 'A',
  '11' : 'A',
  '10' : 'B',
  '9' : 'C',
  '8' : 'C',
  '7' : 'D',
  '6' : 'E',
  '5' : 'Fx',
  '03' : 'Fx',
  '00' : 'F'
  }

  scale7 = {
  '12' : 'A',
  '10' : 'B',
  '7' : 'C',
  '4' : 'D',
  '02' : 'E',
  '00' : 'Fx',
  '-3' : 'F'
  }

  if scale == '13-scale':
    try:
      ECTSGrade = scale13[str(grade)]
    except:
      ECTSGrade = scale13['0' + str(grade)]
  if scale == '7-point':
    try:
      ECTSGrade = scale7[str(grade)]
    except:
      ECTSGrade = scale7['0' + str(grade)]

  return ECTSGrade
```

```
def hireApplicant(skill, pay):

  if skill < 5:
    applicantZone = 'No go'
  else:
    if skill >= 5 and pay > 0.9 * skill + 1:
      applicantZone = 'Too expensive'
    if 5 <= skill < 8 and pay <= 0.9 * skill + 1:
      applicantZone = 'Hire'
    if skill >= 8 and 4 < pay <= 0.9 * skill + 1:
      applicantZone = 'Long term contract'
    if skill >= 8 and pay <= 4:
      applicantZone = 'Unicorn'

  return applicantZone
```

```
def averagedB(M):

  import numpy as np

  removed = []
  for row in M:
    if np.all(row <= 70) == True:
      removed.append(row)
  removed = np.array(removed)
  averageM = np.mean(removed, axis = 0)

  return averageM
```

```
def buildLego(h, l, w):

  bricks = 0
  volume = 0
  while volume + h * l * w <= 1000:
    volume += h * l * w
    bricks += 1

  return bricks
```

```
def sudokuCheck(sudokuBoard):

  import numpy as np

  check = 0
  rows = np.sum(sudokuBoard, axis = 1)
  columns = np.sum(sudokuBoard, axis = 0)
  for row in rows:
    if row != 45:
      check += 1
  for column in columns:
    if column != 45:
      check += 1
  
  for index in range(0, 9, 3):
    if np.sum(sudokuBoard[index:index+3, 0:3]) != 45:
      check += 1
    if np.sum(sudokuBoard[index:index+3, 3:6]) != 45:
      check += 1
    if np.sum(sudokuBoard[index:index+3, 6:9]) != 45:
      check += 1

  return check
```
