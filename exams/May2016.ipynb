```
def eseries(x, N):

  import numpy as np
  import math

  terms = np.arange(N)
  factorials = [math.factorial(term) for term in terms]
  f = sum(x**terms / factorials)

  return f
```

```
def alphaToPhone(alpha):

  conversion = {
  'A' : 2, 'B' : 2, 'C' : 2,
  'D' : 3, 'E' : 3, 'F' : 3,
  'G' : 4, 'H' : 4, 'I' : 4,
  'J' : 5, 'K' : 5, 'L' : 5,
  'M' : 6, 'N' : 6, 'O' : 6,
  'P' : 7, 'Q' : 7, 'R' : 7, 'S' : 7,
  'T' : 8, 'U' : 8, 'V' : 8,
  'W' : 9, 'X' : 9, 'Y' : 9, 'Z' : 9
  }

  phone = ''
  for number in alpha:
    if number in conversion:
      phone += str(conversion[number])
    else:
      phone += number

  return phone
```

```
def genderGuess(name):

  if name[-1] not in ['a','e','i','o','u','y']:
    if name[0] != 'k':
      pFemale = 0.25
    if name[0] == 'k':
      pFemale = 0.69
  if name[-1] in ['a','e','i','o','u','y']:
    if name.count('f') < 2:
      pFemale = 0.87
    if name.count('f') >= 2:
      pFemale = 0.35

  return pFemale
```

```
def birthday(n):

  import math
  P = 1- math.exp(math.lgamma(365 + 1) - math.lgamma(365 - n + 1) - n * math.log(365))

  return P
```

```
def matrixSearch(A, x):

  M, N = np.shape(A)
  i, j = 1, N

  while A[i-1, j-1] != x:

    if A[i-1,j-1] > x:
      j = j-1
    else:
      i = i+1
    
    if i > M or j < 1:
      i = 0
      j = 0
      break
  
  index = np.array([i, j])

  return index
```
