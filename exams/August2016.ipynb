```
def confidence(x):

  import numpy as np
  import math

  m = np.mean(x)
  s = np.std(x, ddof=1)
  interval = 2 * (s / math.sqrt(np.size(x)))
  conf = np.array([m - interval, m + interval])

  return conf
```

```
def weekday(d, m, y):

  import math

  month = {
  1 : 6,
  2 : 2,
  3 : 2,
  4 : 5,
  5 : 0,
  6 : 3,
  7 : 5,
  8 : 1,
  9 : 4,
  10 : 6,
  11 : 2,
  12 : 4
  } 

  week = {
  0 : 'Sun',
  1 : 'Mon',
  2 : 'Tue',
  3 : 'Wed',
  4 : 'Thu',
  5 : 'Fri',
  6 : 'Sat'
  }

  C = month[m]

  w = (d + C + y + math.floor(y / 4)) % 7

  name = week[w]

  return name
```

```
def symmetrize(x):

  import numpy as np
  
  y = x + np.transpose(x) - np.diag(np.diag(x))

  return y
```

```
def voldif(R, n):

  import math

  Vs = (math.pi**(n/2)) / (math.gamma(n/2 +1)) * R**n
  Vc =(2 * R)**n
  Vd = abs(Vs - Vc)

  return Vd
```

```
def stringcompare(string1, string2):

  disSimilarity = 0
  letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
  for letter in letters:
    if letter in string1 and letter not in string2 or letter in string2 and letter not in string1:
      if string1.count(letter) != string2.count(letter):
        disSimilarity += 1

  return disSimilarity
```
