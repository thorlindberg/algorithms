```
def loadBalance(runtime):

  dif = []
  for index in range(len(runtime) - 1):
    c1 = sum(runtime[0:index+1])
    c2 = sum(runtime[index+1:])
    dif.append(abs(c1 - c2))

  for index in range(len(dif)):
    if dif[index] == min(dif):
      k = index + 1
      break
  
  return k
```

```
def nearestColor(r, g, b):

  import numpy as np

  colors = np.array(["White", "Grey", "Black", "Red", "Maroon", "Yellow", "Olive",
  "Lime", "Green", "Aqua", "Teal", "Blue", "Navy", "Fuchsia", "Purple"])
  
  R = np.array([100, 50, 0, 100, 50, 100, 50, 0, 0, 0, 0, 0, 0, 100, 50])
  G = np.array([100, 50, 0, 0, 0, 100, 50, 100, 50, 100, 50, 0, 0, 0, 0])
  B = np.array([100, 50, 0, 0, 0, 0, 0, 0, 0, 100, 50, 100, 50, 100, 50])
  
  distances = np.array([abs(R-r), abs(G-g), abs(B-b)])
  maxes = []
  for column in distances.T:
    maxes.append(np.max(column))
  maxes = np.array(maxes)

  for index in range(len(maxes)):
    if maxes[index] == np.min(maxes):
      colorName = colors[index]
      break

  return colorName
```

```
def climbCategorization(distance, grade):

  if grade > 8:
    if distance < 2:
      return 'Beginner'
    if 2 <= distance < 4:
      return 'Easy'
    if 4 <= distance < 8:
      return 'Medium'
    if 8 <= distance < 12:
      return 'Difficult'
    if 12 <= distance:
      return 'Extreme'
  
  if distance > 16:
    if grade < 1:
      return 'Beginner'
    if 1 <= grade < 2:
      return 'Easy'
    if 2 <= grade < 4:
      return 'Medium'
    if 4 <= grade < 6:
      return 'Difficult'
    if 6 <= grade:
      return 'Extreme'

  if distance * grade < 16:
    return 'Beginner'
  if 16 <= distance * grade < 32:
    return 'Easy'
  if 32 <= distance * grade < 64:
    return 'Medium'
  if 64 <= distance * grade < 96:
    return 'Difficult'
  if 96 <= distance * grade:
    return 'Extreme'
```

```
def romanToValue(roman):

  symbols = {
  'I' : 1,
  'V' : 5,
  'X' : 10,
  'L' : 50,
  'C' : 100,
  'D' : 500,
  'M' : 1000
  }

  total = 0
  maximum = 0
  for symbol in roman[::-1]:

    if symbols[symbol] >= maximum:
      total += symbols[symbol]
      maximum = symbols[symbol]
    if symbols[symbol] < maximum:
      total -= symbols[symbol]
  
  value = total

  return value
```
