```
def normrange(x, r):

  import numpy as np

  if np.max(x) == np.min(x):
    return x

  normalized = np.array([])
  for value in x:
    n = (value - np.min(x)) / (np.max(x) - np.min(x))
    normalized = np.append(normalized, n)
  
  scaled = np.array([])
  for value in normalized:
    s = value * (r[1] - r[0]) + r[0]
    scaled = np.append(scaled, s)

  return scaled
```

```
def readability(text):

  if '.' not in text:
    return 0

  W = len(text.split())

  P = 0
  for letter in text:
    if letter == '.':
      P += 1
      text = text.replace('.','')

  L = 0
  for word in text.split():
    if len(word) > 6:
      L += 1

  LIX = (W / P) + ((L * 100) / W)

  return LIX
```

```
def temperaturemonitor(T):

  import numpy as np
  
  W = np.array([0])

  for index in range(len(T) - 1):
    if T[index] > 5 and T[index + 1] > 5:
      W = np.append(W, 3)
    elif T[index] < 0 and T[index + 1] < 0:
      W = np.append(W, 2)
    elif abs(T[index] - T[index + 1]) > 2:
      W = np.append(W, 1)
    else:
      W = np.append(W, 0)

  return W
```

```
def gamecalculator(goals):

  import numpy as np

  points = 0
  for index in range(len(goals[0,:])):
    if goals[0, index] > goals[1, index]:
      points += 3
    if goals[0, index] == goals[1, index]:
      points += 1
    if goals[0, index] < goals[1, index]:
      points += 0

  GD = np.sum(goals[0,:]) - np.sum(goals[1,:])
  
  result = np.array([points, GD])

  return result
```
