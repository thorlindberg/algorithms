```
def imageCrop(img_in, background):

  import numpy as np

  while True:
    if np.all(img_in[0,:] == background) == True:
      img_in = np.delete(img_in, 0, 0)
    if np.all(img_in[-1,:] == background) == True:
      img_in = np.delete(img_in, -1, 0)
    
    if np.all(img_in[:,0] == background) == True:
      img_in = np.delete(img_in, 0, 1)
    if np.all(img_in[:,-1] == background) == True:
      img_in = np.delete(img_in, -1, 1)
    
    if np.all(img_in[0,:] == background) == False and np.all(img_in[-1,:] == background) == False and np.all(img_in[:,0] == background) == False and np.all(img_in[:,-1] == background) == False:
      break

  img_out = img_in
  return img_out
```

```
def chaos(x, k, T):

  import numpy as np

  seq = np.array([])

  iteration = 0
  while iteration < T:
    x = k * x * (1 - x)
    seq = np.append(seq, x)
    iteration += 1

  return seq
```

```
def threshold(GPA, fail):

  import numpy as np

  sort = np.unique(GPA)

  possible = np.array([])
  for index in range(len(sort) - 1):
    threshold = (sort[index + 1] - sort[index]) / 2 + sort[index]
    possible = np.append(possible, threshold)

  Fscores = np.array([])
  for X in possible:

    count = 0
    for index in range(len(GPA)):
      if GPA[index] <= X and fail[index] == 1:
        count += 1

    score = 2 * ( count / ( len(GPA[GPA <= X]) + len(fail[fail == 1]) ) )
    Fscores = np.append(Fscores, score)

  for index in range(len(Fscores)):
    if Fscores[index] == np.max(Fscores):
      X = possible[index]
      break

  return X
```

```
def checkParentheses(str):

  matching = {
  '(' : ')',
  '[' : ']',
  '{' : '}'
  }

  stack = []
  for index in range(len(str)):
    if str[index] in '([{':
      stack.append(str[index])
    if str[index] in ')]}':
      if stack and str[index] == matching[stack[-1]]:
        stack.pop()
      else:
        return index + 1
  
  if len(stack) == 0:
    return 0
  else:
    return len(str) + 1
```
