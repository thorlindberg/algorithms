```
def salePrice(prices):

  prices = sorted(prices, reverse=True)

  total = 0
  for index in range(0,len(prices),2):
    total += prices[index]

  return total
```

```
def zeroCrossing(signal):

  sign = ''
  count = -1
  for number in signal:
    if number < 0 and sign != '-':
      sign = '-'
      count += 1
    if number >= 0 and sign != '+':
      sign = '+'
      count += 1

  return count
```
