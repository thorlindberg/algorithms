```
def rgb2hue(R, G, B):

  dif = max(R, G, B) - min(R, G, B)
  
  if R == max(R, G, B):
    H = 60 * ((G - B) / dif)
  if G == max(R, G, B):
    H = 120 + 60 * ((B - R) / dif)
  if B == max(R, G, B):
    H = 240 + 60 * ((R - G) / dif)
  
  if H < 0:
    H += 360

  return H
```

```
def cardValidation(cardnumber):

  table = {
  '0' : '0',
  '1' : '2',
  '2' : '4',
  '3' : '6',
  '4' : '8',
  '5' : '1',
  '6' : '3',
  '7' : '5',
  '8' : '7',
  '9' : '9'
  }

  conversion = []
  for index in range(0, len(cardnumber) - 1, 2):
    conversion.append(int(table[cardnumber[index]]))
    conversion.append(int(cardnumber[index + 1]))
  checksum = sum(conversion)

  return checksum
```

```
def polygonPerimeter(x, y):

  import math

  P = 0
  for index in range(len(x) - 1):
    xn = x[index] - x[index + 1]
    yn = y[index] - y[index + 1]
    P += math.sqrt(xn**2 + yn**2)
  
  xn = x[-1] - x[0]
  yn = y[-1] - y[0]
  P += math.sqrt(xn**2 + yn**2)

  return P
```

```
def rotateScale(coordinates, center, theta, scale):

  import numpy as np
  import math

  rotation = np.array([
  [math.cos(theta), -math.sin(theta)],
  [math.sin(theta), math.cos(theta)]
  ])
  
  center = np.matrix(center).T

  newCoordinates = rotation * (coordinates - center) * scale + center

  return newCoordinates
```

```
import numpy as np
import math
rotateScale(np.array([[1,3,3,4,7,6,1],[1,1,2,3,3,5,5]]), np.array([3,1]), -math.pi/3, 2)
```

```
def costliestCar(maxPrice):

  BasePrice = 159000
  TrimPrice = np.array([0, 22000, 44000])
  ExtraPrice = np.array([4000, 8000, 13000, 7000])
  TrimString = np.array(["Access", "Comfort", "Sport"])
  ExtraString = np.array(["Cruise control", "Air conditioning", "Alloy wheels", "Chrome spoiler"])

  costliestPrice = 0
  for trim in range(3):
    for extra in range(2**4):
      extra_indicator = np.floor(extra / 2**np.arange(4)) % 2 == 1
      price = BasePrice + TrimPrice[trim] + np.sum(ExtraPrice[extra_indicator])
      if (price > costliestPrice) and (price <= maxPrice):
        costliestPrice = price
        costliestTrim = trim
        costliestExtra = extra_indicator	   

  carSpecification = ", ".join(np.append(TrimString[costliestTrim], ExtraString[costliestExtra]))

  return carSpecification
```
