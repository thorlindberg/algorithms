```
def computeProjection(a):

  b = np.ones(np.size(a))

  projection = np.dot(a, b) / np.linalg.norm(a)**2 * a

  return projection
```

```
def boxArea(boxCorners, area):

  if area == 'Box1':
    A = (boxCorners[1] - boxCorners[0]) * (boxCorners[5] - boxCorners[4])
  elif area == 'Box2':
    A = (boxCorners[3] - boxCorners[2]) * (boxCorners[7] - boxCorners[6])
  elif area == 'Intersection':
    A = np.dot(
    max(0, min(boxCorners[1], boxCorners[3]) - max(boxCorners[0], boxCorners[2])),
    max(0, min(boxCorners[5], boxCorners[7]) - max(boxCorners[4], boxCorners[6]))
    )
  elif area == 'Union':
    A1 = (boxCorners[1] - boxCorners[0]) * (boxCorners[5] - boxCorners[4])
    A2 = (boxCorners[3] - boxCorners[2]) * (boxCorners[7] - boxCorners[6])
    AI = np.dot(
    max(0, min(boxCorners[1], boxCorners[3]) - max(boxCorners[0], boxCorners[2])),
    max(0, min(boxCorners[5], boxCorners[7]) - max(boxCorners[4], boxCorners[6]))
    )
    A = A1 + A2 - AI
  return A

import numpy as np
print(boxArea(np.array([5,20,14,25,12,23,5,17]), "Intersection"))
```

```
def fillSudokuRow(sudokuRow):

  sudokuRow[sudokuRow == 0] = np.sum(np.arange(1, 10)) - np.sum(sudokuRow)

  return sudokuRow

import numpy as np
print(fillSudokuRow(np.array([9, 4, 0, 1, 5, 7, 2, 3, 8])))
```
