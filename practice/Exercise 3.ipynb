```
import numpy as np

def pH2Category(pH):

  category = 'pH out of range'
  if 0 <= int(pH) <= 2:
    category = 'Strongly acidic'
  if 3 <= int(pH) <= 5:
    category = 'Weakly acidic'
  if 6 <= pH <= 8:
    category = 'Neutral'
  if 9 <= np.ceil(pH) <= 11:
    category = 'Weakly basic'
  if 12 <= np.ceil(pH) <= 14:
    category = 'Strongly basic'

  return category

print(pH2Category(2.3)) 
```

```
import numpy as np

def acuteAngle(v1, v2):

  angle = np.arccos(np.dot(v1, v2))

  if angle > np.pi / 2:
    theta = np.pi - angle
  else:
    theta = angle

  return theta

print(acuteAngle(np.array([-4.0/5.0, 3.0/5.0]), np.array([20.0/29.0, 21.0/29.0])))
```

```
def gravitationalPull(x):

  g0 = 9.82
  R = 6.371 * 10**6

  if R <= x:
    g = g0 * (R**2 / x**2)
  if 0 <= x < R:
    g = g0 * (x / R)

  return g

import numpy as np
print(gravitationalPull(1.78e6))
```
