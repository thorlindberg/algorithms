```
def removeIncomplete(id):

  idComplete = np.array([])
  for index in id:
    if int(index) + 0.1 in id and int(index) + 0.2 in id and int(index) + 0.3 in id:
      idComplete = np.append(idComplete, index)

  return idComplete

import numpy as np
print(removeIncomplete(np.array([1.3, 2.2, 2.3, 4.2, 5.1,
3.2, 5.3, 3.3, 2.1, 1.1, 5.2, 3.1])))
```

```
def bacteriaGrowth(n0, alpha, K, N):

  tN = 0
  nt = n0
  while nt < N:
    nt = (1 + alpha * (1 - nt / K)) * nt
    tN += 1

  return tN

print(bacteriaGrowth(100.0, 0.4, 1000.0, 500.0)) 
```

```
def fermentationRate(measuredRate, lowerBound, upperBound):

  rates = np.array([])

  for rate in measuredRate:
    if lowerBound < rate < upperBound:
      rates = np.append(rates, rate)

  averageRate = np.mean(rates)

  return averageRate

import numpy as np
print(fermentationRate(np.array([20.1, 19.3, 1.1, 18.2, 19.7, 121.1, 20.3, 20.0]), 15, 25))
```

```
def clusterAnalysis(reflectance):

  class1 = np.array(reflectance[::2])
  class2 = np.array(reflectance[1::2])
  clusterAssignments = np.empty(len(reflectance))
  clusterAssignments[::2] = 2
  clusterAssignments[1::2] = 1

  while True:

    mean1 = np.mean(class1)
    mean2 = np.mean(class2)

    compareCluster = np.array([])
    class1 = np.array([])
    class2 = np.array([])

    for measurement in reflectance:
      if abs(measurement - mean1) <= abs(measurement - mean2):
        class1 = np.append(class1, measurement)
        compareCluster = np.append(compareCluster, 1)
      if abs(measurement - mean1) > abs(measurement - mean2):
        class2 = np.append(class2, measurement)
        compareCluster = np.append(compareCluster, 2)

    if np.array_equal(clusterAssignments, compareCluster) == True:
      break
    else:
      clusterAssignments = compareCluster

  return clusterAssignments

import numpy as np
print(clusterAnalysis(
np.array([1.7, 1.6, 1.3, 1.3, 2.8, 1.4, 2.8, 2.6, 1.6, 2.7])
))
```
