```
def convertTemperature(T, unitFrom, unitTo):

  conversion = {
  'Celsius' : { 'Fahrenheit' : 1.8 * T + 32, 'Kelvin' : T + 273.15 },
  'Fahrenheit' : { 'Celsius' : (T - 32) / 1.8, 'Kelvin' : (T + 459.67) / 1.8 },
  'Kelvin' : { 'Celsius' : T - 273.15, 'Fahrenheit' : 1.8 * T - 459.67 }
  }

  T = conversion[unitFrom][unitTo]

  return T

print(convertTemperature(50.0, "Fahrenheit", "Celsius"))
```

```
def thermoEquilibrium(N, r):

  NL = N
  NR = 0
  t = 0

  while NL != NR:

    t += 1
    PLR = NL / N
    PRL = 1 - PLR

    if 0 <= r[t-1] <= 1:
      if r[t-1] <= PLR:
        NL -= 1
        NR += 1
      else:
        NL += 1
        NR -= 1
    
    if t == len(r):
      t = 0
      break

  return t

import numpy as np
print(thermoEquilibrium(2.0, np.array([0.16, 0.04, 0.72,
0.09, 0.17, 0.60, 0.26, 0.65, 0.69, 0.74, 0.45, 0.61,
0.23, 0.37, 0.15, 0.83, 0.61, 1.00, 0.08, 0.44])))
```
